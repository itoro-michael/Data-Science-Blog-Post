```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline

df = pd.read_csv('2017-survey-results-public.csv')
df.head()
```

```python
df.describe()
```

```python
#The number of rows and columns
print("The number of rows:",df.shape[0], "\nThe number of columns:", df.shape[1])
```

```python
# Show the proportion of the WorkRemote categories in the 2017  Dataset
val_count = df.HomeRemote.value_counts() #Provide a pandas series

# The below should be a bar chart
val_count = val_count/df.shape[0]
val_count = val_count.sort_values()
val_count.plot(kind="barh", color='#1F77B4', title='Remote Work Categories in Dataset', figsize=(10,5));
for y, x in enumerate(val_count.values):
    plt.annotate(str(round(x,2)), xy=(x, y), va='center')
```
